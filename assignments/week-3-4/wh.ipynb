```python
#Combining 'Last Name' and 'First Name' into a 'Visitor' column
if 'Last Name' in df.columns and 'First Name' in df.columns:
    df['Visitor'] = df['Last Name'] + ", " + df['First Name'] + " " + df['Middle Initial']
else:
    print("The 'Last Name' or 'First Name' column is missing.")

print(df[['Last Name', 'First Name', 'Visitor']].head())
```

```python
#Combining 'Visitee Last Name' and 'Visitee First Name' into a 'Visitee' column
if 'Visitee Last Name' in df.columns and 'Visitee First Name' in df.columns:
    df['Visitee'] = df['Visitee Last Name'] + ", " + df['Visitee First Name']
else:
    print("The 'Last Name' or 'First Name' column is missing.")

print(df[['Visitee Last Name', 'Visitee First Name', 'Visitee']].head())
```

```python
#Combining 'Caller Last Name' and 'Caller First Name' into a 'Caller' column
if 'Caller Last Name' in df.columns and 'Caller First Name' in df.columns:
    df['Caller'] = df['Caller Last Name'] + ", " + df['Caller First Name']
else:
    print("The 'Last Name' or 'First Name' column is missing.")

print(df[['Caller Last Name', 'Caller First Name', 'Caller']].head())
```

```python
#Dropping unecessary columns
columns_to_drop = [
    'UIN',
    'BDGNBR',
    'Access Type',
    'POA', 
    'TOD', 
    'POD', 
    'Appointment Cancel Date',
    'POST',
    'CALLER_ROOM', 
    'RELEASEDATE', 
    'Caller Room', 
    'Release Date'
]

df = df.drop(columns=columns_to_drop, errors='ignore')
```

```python
#Keeping only a school year's worth of data.
df['Appointment Start Date'] = pd.to_datetime(df['Appointment Start Date'])

start_date = '2022-08-01'
end_date = '2023-06-30'

df = df[(df['Appointment Start Date'] >= start_date) & (df['Appointment Start Date'] <= end_date)]
```

```python
#Extracting month data.
df['Appointment Start Date'] = pd.to_datetime(df['Appointment Start Date'])
df['YearMonth'] = df['Appointment Start Date'].dt.to_period('M')
```

```python
#Grouping by 'YearMonth' and summing 'Total People'.
monthly_visits = df.groupby('YearMonth')['Total People'].sum().reset_index()
```

```python
#Converting 'YearMonth' back to datetime for plotting.
monthly_visits['YearMonth'] = monthly_visits['YearMonth'].dt.to_timestamp()

#Plotting a line graph.
plt.figure(figsize=(12, 6))
plt.plot(monthly_visits['YearMonth'], monthly_visits['Total People'], marker='o', linestyle='-', color='#ff5757')

plt.ylim(bottom=0)
plt.xlabel("Month")
plt.ylabel("Number of Attendees")
plt.title("Tour Attendees by Month")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()

plt.show()
```

```python
#Extracting the day of the week (0=Monday, 6=Sunday)
df['Day of Week'] = df['Appointment Start Date'].dt.day_name()

day_counts = df['Day of Week'].value_counts().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
```

```python
# -------------------------
# 6. Bubble Chart: Repeat Visitor Influence
# -------------------------
repeat = df["Visitor"].value_counts().reset_index()
repeat.columns = ["Visitor", "count"]

plt.figure(figsize=(12,6))
plt.scatter(repeat.index, repeat["count"],
            s=repeat["count"]*10, alpha=0.6)
plt.title("Visitor Influence (Repeat Visits)")
plt.xlabel("Visitor Index")
plt.ylabel("Visit Count")
plt.tight_layout()
plt.show()
```

```python
# Filter for tours, process date, aggregate, and plot
# 1. Filter for Tours: Use the CALLER_NAME_FIRST column and filter for 'TOUR'
tour_df = df[df['Caller First Name'] == 'TOUR'].copy()

# 2. Process Date: Convert APPT_START_DATE to datetime objects
# We use errors='coerce' to handle any inconsistent date formats, as hinted in the original PDF script analysis.
tour_df['APPT_DATE'] = pd.to_datetime(tour_df['Appointment Start Date'], errors='coerce').dt.date

# Drop rows where date conversion failed (should be minimal)
tour_df.dropna(subset=['APPT_DATE'], inplace=True)

# 3. Aggregate: Count the number of tours per day
daily_tours = tour_df.groupby('APPT_DATE').size()
daily_tours_df = daily_tours.reset_index(name='Tour Count')

# 4. Plot: Create a line plot for daily tour counts
plt.figure(figsize=(12, 6))
plt.plot(daily_tours.index, daily_tours.values, marker='o', linestyle='-', color='#003366')

plt.title('Daily White House Tour Count', fontsize=16)
plt.xlabel('Date )', fontsize=12)
plt.ylabel('Number of Tours', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the plot
plt.show()

print(daily_tours_df.head())
print(daily_tours_df.info())
```

```python
#Extract time data.
df['Hour'] = df['Appointment Start Date'].dt.hour
hourly_counts = df['Hour'].value_counts().sort_index()
```

```python
#Plotting a step chart.
hourly_counts = df['Hour'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.step(hourly_counts.index, hourly_counts.values, where='mid', marker='o', linestyle='-')

plt.xlabel("Hour of the Day")
plt.ylabel("Number of Appointments")
plt.title("Distribution of Appointments by Hour")
plt.grid(axis='y')
plt.tight_layout()

plt.show()
```

```python
#Adding hour category to simplify coloring of graph.
df['Hour Category'] = pd.cut(
    df['Hour'],
    bins = [11, 16, 22],
    labels = ['School Hours', 'After School Hours'],
    right = False)

#Plotting a stacked bar chart.
hourly_day_counts = df.groupby(['Day of Week', 'Hour Category']).size().unstack(fill_value=0)

ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
hourly_day_counts = hourly_day_counts.reindex(ordered_days)

colors = ['#2d4578', '#ff5757']

plt.figure(figsize=(12, 8))
hourly_day_counts.plot(kind='bar', stacked=True, color=colors, edgecolor='black')

plt.xlabel("Day of the Week")
plt.ylabel("Number of Tours")
plt.title("Tour Times Across Days of the Week")
plt.xticks(rotation=45)
plt.legend(title='Hour of Day', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y')
plt.tight_layout()

plt.show()
```
